### Verwendung von Variablen - Repräsentation von Buchstaben
#### Die ASCII - Tabelle
ASCII steht für *American Standard Code for Information Interchange*.

```python
buchstaben = list("abcdefghijklmnopqrstuvwxyz")
```

#### Imports

```python
import pandas as pd
import numpy as np
```

#### ASCII Table

```python
ascii_werte = [ord(buchstabe) for buchstabe in buchstaben]
binaer_ascii_werte = [bin(wert) for wert in ascii_werte]
hex_ascii_werte = [hex(wert) for wert in ascii_werte]
```

```python
df = pd.DataFrame({"Buchstaben": buchstaben, 
                   "ASCII-Werte": ascii_werte,
                   "Binär-ASCII-Werte": binaer_ascii_werte, 
                   "Hexadezimal-ASCII-Werte": hex_ascii_werte})
df
```

Bisher wurden die ASCII-Wertzuordnungen von "a" bis "z" veranschaulicht. Nun schauen wir und den gesamten Zuordnungsbereich von einem Byte (bzw. 8 Bit an). Mit 8 Bit lassen sich $2^8 = 256$. Ursprünglich bestand der ASCII sogar nur aus 7 Bit ($2^7 = 128$). 

```python
2**8, 2**7
```

Nun, können wir uns einmal alle im ASCII abgebildeten Zeichen anschauen. Hierzu durchlaufen wir alle Zahlenwerte und geben uns das entsprechende Zeichen mit Hilfe der $\texttt{chr}$-Funktion aus:

```python
wertebereich = range(0, 2**7)
alle_zeichen = [chr(wert) for wert in wertebereich]
```

```python
print(alle_zeichen)
```

```python
tabelle = pd.DataFrame({"Dezimal": wertebereich,
                        "Hex": [hex(wert) for wert in wertebereich],
                        "Binär": [bin(wert) for wert in wertebereich],
                        "Zeichen": alle_zeichen})
```
